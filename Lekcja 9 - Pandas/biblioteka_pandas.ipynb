```python
# 1. Wczytaj dane z pliku Gigathon_Wyniki.xlsx udostępnionego przez trenera
import pandas as pd

df = pd.read_excel("/content/Gigathon_Wyniki.xlsx") # ścieżka do pliku
```

```python
# 2. Wyświetl zawartość pliku jako dataframe
df
```

```python
# 3. Wyświetl 5 pierwszych, a następnie 5 ostatnich wpisów w pliku

print(df.head(5))

print(df.tail(5))
```

```python
# 4. Dodaj kolumnę “Wynik procentowy” która będzie zawierała informację o wyniku przeliczonym na punkty.
df["Wynik procentowy"] = df["Punktacja"]/120 * 100
df
```

```python
# 5. Przefiltruj listę uczestników tak, aby zostały tylko osoby które nadesłały zadania konkursowe

# Do usunięcia rekordów z wartością NaN służy metoda .dropna().
# Parametr inplace informuje że usuwamy dane z oryginalnego
# dataframe, domyślnie metoda zwraca kopię dataframe z usuniętymi
# rekordami.

df.dropna(inplace=True)
df
```

```python
# 6. Posortuj uczestników od najlepszego wyniku do najsłabszego

# Do sortowania wartości służy metoda .sort_values().
# Parametry:
# 'by' - według której kolumny odbywa się sortowanie,
# 'ascending' - kolejność sortowania (rosnąca lub malejąca).

df.sort_values(by="Punktacja", ascending=False, inplace=True)
df
```

```python
# 7. Dodanie kolumny zakwalifikowany,
# w której przypisana zostanie wartość True dla osób które uzyskały ponad 70% możliwych do uzyskania punktów,
# reszta osób powinna mieć przypisaną wartość False

df["Zakwalifikowany"] = df["Wynik procentowy"] > 70
df
```

```python
# 8. Przygotuj zestawienie które pogrupuje uczestników względem zakwalifikowania się do kolejnego etapu i wieku.
# Dla każdej grupy zlicz ile osób mieści się w tym przedziale, jaki był najniższy i największy wynik w danym przedziale,
# oraz jaka była średnia i mediana.

df.groupby(["Zakwalifikowany", "Wiek"])[["Punktacja", "Wynik procentowy"]].agg(["count", "min", "max", "mean", "median"])
```

```python
# 9. Rozbij tabelę na dwie mniejsze tabele - zakwalifikowani oraz niezakwalifikowani. Zapisz je do oddzielnych plików .xlsx

# Zakwalifikowani:
df_zakwalifikowani = df[df["Zakwalifikowany"]]
df_zakwalifikowani.drop("Zakwalifikowany", axis=1, inplace = True)
df_zakwalifikowani.to_excel("zakwalifikowani.xlsx")
df_zakwalifikowani
```

```python
# Niezakwalifikowani
df_niezakwalifikowani = df[~df["Zakwalifikowany"]]
df_niezakwalifikowani.drop("Zakwalifikowany", axis=1, inplace = True)
df_niezakwalifikowani.to_excel("niezakwalifikowani.xlsx")
df_niezakwalifikowani
```
